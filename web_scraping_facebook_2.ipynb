```python
import requests
from facebook_scraper import get_posts
from facebook_scraper import get_profile

import pandas as pd
from datetime import datetime
```

```python
def extraer_facebook(unique_name, pages): #recibe nombre unico en facebook y cantidad de pags a scrapear (10post por pag)
    posts0 = []
    for post in get_posts(unique_name, pages=pages, credentials= ('perlatrabajo', 'Violin.00'),
                           options={"posts_per_page": 15, "comments": True, "reactors": True}):
        posts0.append(post)
    return posts0
```

```python
def genera_tabla_final(df):
    #tabla_general = df[['username','post_id','time','post_text','likes','comments','shares',
    #                 'video','comments_full', 'reactions','image']]
    tabla_general = df
    tabla_general1 = pd.DataFrame()
    ind=0
    for i in range(len(tabla_general)):
        if len(tabla_general.comments_full[i]) != 0:
            for c in range(len(tabla_general.comments_full[i])):
                tabla_general1.loc[ind,'comment_id']= tabla_general.comments_full[i][c]['comment_id']
                tabla_general1.loc[ind,'commenter_name']= tabla_general.comments_full[i][c]['commenter_name']
                tabla_general1.loc[ind,'comment_text']= tabla_general.comments_full[i][c]['comment_text']
                tabla_general1.loc[ind,'comment_time']= tabla_general.comments_full[i][c]['comment_time']
                tabla_general1.loc[ind,'comment_reaction_count']= tabla_general.comments_full[i][c]['comment_reaction_count']
                tabla_general1.loc[ind,'username_post']= tabla_general.username[i]
                tabla_general1.loc[ind,'post_id'] = tabla_general.post_id[i]
                tabla_general1.loc[ind,'time_post'] = tabla_general.time[i]
                tabla_general1.loc[ind,'post_text'] = tabla_general.post_text[i]
                tabla_general1.loc[ind,'likes_post']= tabla_general.likes[i]
                #tabla_general1.loc[ind,'reactions']= (tabla_general.reactions[ind])
                tabla_general1.loc[ind,'comments_post']= tabla_general.comments[i]
                tabla_general1.loc[ind,'shares_post'] = tabla_general.shares[i]
                tabla_general1.loc[ind,'video_post'] = tabla_general.video[i]
                tabla_general1.loc[ind,'image_post'] = tabla_general.image[i]
                #tabla_general1.loc[ind,'video_watches'] = tabla_general.video_watches[i]
                ind += 1
        else:
                tabla_general1.loc[ind,'comment_id']= 'None'
                tabla_general1.loc[ind,'commenter_name']= 'None'
                tabla_general1.loc[ind,'comment_text']= 'None'
                tabla_general1.loc[ind,'comment_time']= 'None'
                tabla_general1.loc[ind,'comment_reaction_count']= 'None'
                tabla_general1.loc[ind,'username_post']= tabla_general.username[i]
                tabla_general1.loc[ind,'post_id'] = tabla_general.post_id[i]
                tabla_general1.loc[ind,'time_post'] = tabla_general.time[i]
                tabla_general1.loc[ind,'post_text'] = tabla_general.post_text[i]
                tabla_general1.loc[ind,'likes_post']= tabla_general.likes[i]
                #tabla_general1.loc[ind,'reactions']= (tabla_general.reactions[ind])
                tabla_general1.loc[ind,'comments_post']= tabla_general.comments[i]
                tabla_general1.loc[ind,'shares_post'] = tabla_general.shares[i]
                tabla_general1.loc[ind,'video_post'] = tabla_general.video[i]
                tabla_general1.loc[ind,'image_post'] = tabla_general.image[i]
                #tabla_general1.loc[ind,'video_watches'] = tabla_general.video_watches[i]
                ind += 1        

    return tabla_general1
```
